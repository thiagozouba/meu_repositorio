```python
# Importações de Biblioecas e alguns comandos para análise de dados

import pandas as pd


df = pd.read_csv('dataset_trabalho.csv')

df.info() # Descreve as estruras do DataFrame

df.describe() # Descreve valores para fins estatísticos

df.head() # Exibe as primeiras linhas

df.tail() # Exibe as últimas linhas

df.isnull() # Exibe os valores nulos

df.nunique()    # Exibe o número de valores únicos em cada coluna do DataFrame



```

```python
# Verificação inicial
df.info()
df.isnull().sum()

# Tratamento
df['Rating'].fillna('Desconhecido', inplace=True)
df['User_Score'] = pd.to_numeric(df['User_Score'], errors='coerce')
df = df.dropna()

# Remover duplicatas
df = df.drop_duplicates()
```

```python
# Mostrar as tipagens de dados
df.dtypes
```

```python
#10 Perguntaas para análise de dados:
```

```python
# 1 -Quais são os 20 jogos mais vendidos?

df[['Name', 'Global_Sales', 'Year_of_Release']].sort_values(by='Global_Sales', ascending=False).head(20)
```

```python
# 2 - Qual a quantidade de jogos por gênero?

df['Genre'].value_counts()
```

```python
# 3- Qual o ano que teve mais vendas?

df.groupby('Year_of_Release')['Global_Sales'].sum().sort_values(ascending=False)
```

```python
# 4 -Qual a Publisher que mais vendeu?

df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(15)
```

```python
# 5 - Quais as regiões que mais venderam?

df[['NA_Sales', 'EU_Sales', 'JP_Sales']].sum()
```

```python
# 6 - Qual a média de vendas globais por Publisher?

df.groupby('Publisher')['Global_Sales'].mean().sort_values(ascending=False)
```

```python
# 7 - Quais os jogos que tiveram mais vendas na região da América do Norte e Japão?

df['Diferenca'] = abs(df['NA_Sales'] - df['JP_Sales'])

df[['Name', 'Diferenca']].sort_values(by='Diferenca', ascending=False).head(10)

```

```python
# 8 - Quantos jogos possuem nota crítica acima de 90?

df[df['Critic_Score'] > 90]['Name'].count()

```

```python
# 9 - Qual o total de vendas da franquia Mario, seu ano de lançamento e quais as regiões que mais venderam? 

mario = df[df['Name'].str.contains('Mario', case=False)] # Filtrar jogos do Mario
```

```python
mario.groupby('Name')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Year_of_Release']].sum() # Mostrar os nomes dos Jogos
```

```python
mario['Global_Sales'].sum() # Total de vendas da franquia
```

```python
mario[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum() # Regiões que mais venderam
```

```python
# 10 - A publisher CAPCOM, vendeu mais em qual região? E quais os jogos mais vendidos?

# Filtrar :

capcom_sales = df[df['Publisher'] == 'Capcom']

capcom_sales

```

```python
# Ver em qual região vendeu mais:

capcom_sales[['NA_Sales', 'EU_Sales', 'JP_Sales']].sum()
```

```python
# Ver jogos mais vendidos por ano de lançamento:
    
capcom_sales[['Name', 'Global_Sales', 'Year_of_Release']].sort_values(by='Global_Sales', ascending=False).head(15)
```
